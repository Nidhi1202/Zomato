##### the no. of null rows in rate column are highest so we begin with cleaning that column  first

```python
df['rate'].unique()
```

##### we can clearly see there are NEW and nan values which are useless and needs to be removed. along with this, we can also remove 5 from the denominator for a clear view


```python
def ratingcheck(value):
    if value in ['NEW', '-']:
        return np.nan
    else:
        value = str(value).split('/')
        return float(value[0])
```

##### for cleaning rate column, we cannot drop those columns as no. of NON NULL rows are more than 50 percent. Here we are defining a function where if NEW or - values ar encountered, we fill those with nan. for removing denominator, we use split funtion and retain the value [0] by removing denominator. We return the float value and then apply the defined funtion to rate column below

```python
df['rate'] = df['rate'].apply(ratingcheck)
```

```python
pd.isnull(df['rate']).sum()
```

##### .isnull() is a pandas series that checks each value in selected column for null or missing values and resturns a boolean serires. .sum() is used to calculate the sum of the boolean series. by using this command we can get the value of missing values for column rate.

```python
df['rate'] = df['rate'].fillna(df['rate'].mean())
```

```python
pd.isnull(df['rate']).sum()
```

##### in this way, we first calculate the mean of all non-null values in rate using .mean() and fill the mean values using .fillna() method and we are directly returning the value in df['rate']. If we will not be returning value in df['rate'] column, it will only give us a new table and not update the main table. We can also use inplace funtion where on returning TRUE=1 means that the updation is done in the main row itself. Now when we check again, the null values in rate column are zero.

```python
df.info()
```

##### Now we can see that there are many values here who have ',' between them which makes it look a bit unclean. So our goal is to remove this comma and for this we will create another funtion and apply on the column.

```python
def comma(value):
    if isinstance(value, str):
        value = value.replace(',', '')
    return float(value)
```

```python
df['CostPer2'] = df['CostPer2'].apply(comma)
```

```python
df['CostPer2'].unique()
```

##### now that we have clean columns like name	online_order book_table rate votes CostPer2, we are onto cleaning rest_type column. For that, we will first check how many unique values does rest_type contain/

```python
rest_types = df['rest_type'].groupby(df['rest_type']).size().sort_values(ascending=False)
```

```python
rest_types
```

##### this command selects the 'rest_type' column from the DataFrame df, groups the DataFrame by the values and calculates the size or count by sorts the resulting counts in descending order. By specifying ascending=False, the values are automatically sorted from highest to lowest.

```python
rt1000 = rest_types.loc[rest_types < 1000]
```

```python
rt1000
```

##### rest_types refers to the Series obtained from the previous step, while rest_types < 1000 creates a boolean mask that checks each value in the rest_types Series and returns True for values that are less than 1000, and False if they are not. rest_types.loc filters the Series to include only those values that are less than 1000.

```python
def rt(loc):
    if loc not in rt1000:
        return loc
    else:
        return 'others'
```

```python
df['rest_type'] = df['rest_type'].apply(rt)
```

```python
rest_types
```

```python
df['rest_type'].value_counts()
```

##### As we can see, we created a function called rt, which created a new cluster of items under 1000 named rt1000 and wass named others. so now, we can see as above, there is a new cluster called above which contains all the values less than 1000 and the dataset now looks much clean. Now that are rest_types column is clean, we could check to clean the location column.

```python
location = df['location'].groupby(df['location']).size().sort_values(ascending=False)
```

##### Just like rest_types, we can see that there are many locations which have very few restaurants as compared to others. So we would do the same as we did for rest_type, we will create another cluster named others and will input these locations having less than 500 restaurants in the area into others and apple to loc column.

```python
lt500 = location[location<500]
```

```python
def lt(loca):
    if loca not in lt500:
        return loca
    else:
        return 'others'
```

```python
df['location'] = df['location'].apply(lt)
```

```python
df['location'].value_counts()
```

##### Now we can see that the similar to rest_type, the location column is cleaned. We are left with cuisines and type columns. Similar to rest_types and location columns, we will first check if there are many unique values present, if there are, then we woul apply similaar process by clustering the data into others.

```python
df['cuisines'].value_counts()
```

```python
cuisines = df['cuisines'].groupby(df['cuisines']).size().sort_values(ascending=False)
```

```python
ct400 = cuisines[cuisines<400]
```

```python
def ct(cui):
    if cui not in ct400:
        return cui
    else:
        return 'others'
```

```python
df['cuisines'] = df['cuisines'].apply(ct)
```

```python
df['cuisines'].value_counts()
```

## DATA VISUALISATION 

```python
plt.figure(figsize=(16, 10))
sns.countplot(x='location', data=df)
plt.xticks(rotation=90)
```

#### Line plt.figure(figsize=(16, 10)) creates a new blank figure with a larger size of 16 inches width and 10 inches height. While sns.countplot(x='location', data=df) creates a bar plot called a "count plot" using Seaborn's countplot function. It counts the occurrences of each unique value in the 'location' column of the DataFrame df. Lastly, we use plt.xticks(rotation=90)for rotating the x-axis tick labels by 90 degrees. The x-axis tick labels represent the unique values of the 'location' column. By rotating them, we make them vertical instead of horizontal. By this, we can see that if i want to open a restaurant, i would certinly avoid opening at places like BTM or HSR, since they already have a huge chunk of restaurants. Instead, I would consider opening restaurants at Shanti nagar or Church street, since they have very less restaueants

```python
df1 = df.groupby(['location', 'online_order'])['name'].count().unstack(fill_value=0)
```

```python
df1
```

##### here, we are creating a pivot table, where we can see which location has online order facility. It is done by it counting the number of occurrences of each combination of location and online order category using the 'name' column. The result is a Series with the count values. While .unstack(fill_value=0) converts the grouped data into a pivot table format.  Any missing combinations are filled with zeros using the fill_value=0 parameter.

```python
plt.figure(figsize=(15, 8))
df1.plot.bar()
plt.show()
```

##### By the above command, we are inputting the data of pivot table in the bar graph. Now, if we want to open a restaurant with online order facility, we would open it at locations like Lavelle road, as there are very less restaurants and majority of them do not have the online order facility.
